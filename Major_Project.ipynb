```python
# -- installs --
!pip -q install sentence-transformers transformers accelerate faiss-cpu pypdf google-generativeai cohere

import os, re, json, time, torch, faiss, numpy as np
from pathlib import Path
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from typing import List, Dict

# ---------------------------
# 🔑 API KEYS (set here or Runtime → Env)
# ---------------------------
os.environ["GOOGLE_API_KEY"]  = "AIzaSyAq_t93GdSHc_BqvE756ntoug37sjKdYdc"
os.environ["COHERE_API_KEY"]  = "ygNgfxHdydrQSJUnBWtr6lHhaLYJYfZ5gPa4yF71"

if not os.getenv("GOOGLE_API_KEY"):
    raise RuntimeError("GOOGLE_API_KEY not set")
if not os.getenv("COHERE_API_KEY"):
    raise RuntimeError("COHERE_API_KEY not set")

# Models (Gemini must include "models/" prefix)
GEMINI_MODEL    = "models/gemini-2.5-flash"          # primary
GEMINI_FALLBACK = "models/gemini-1.5-flash-latest"   # fallback
COHERE_MODEL    = "command-a-03-2025"

# ---------------------------
# 📂 Paths
# ---------------------------
ROBERTO_MODEL_DIR = "/content/drive/MyDrive/MajorProj/roberta_model_best"
BASE_DIR          = "/content/drive/MyDrive/data"
TXT_DIR           = f"{BASE_DIR}/processed"
NEW_KNOW_FILE     = f"{TXT_DIR}/new_knowledge.txt"

Path(TXT_DIR).mkdir(parents=True, exist_ok=True)
Path(BASE_DIR).mkdir(parents=True, exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"

# ===========================================================
# 1) PROMPT INJECTION DETECTOR (RoBERTa — local)
# ===========================================================
tokenizer_guard = AutoTokenizer.from_pretrained(ROBERTO_MODEL_DIR, local_files_only=True)
model_guard     = AutoModelForSequenceClassification.from_pretrained(ROBERTO_MODEL_DIR, local_files_only=True).to(device).eval()

@torch.inference_mode()
def score_injection(text: str) -> float:
    x = tokenizer_guard(text, return_tensors="pt", truncation=True, max_length=512).to(device)
    logits = model_guard(**x).logits
    probs  = torch.softmax(logits, dim=-1)[0].tolist()
    return float(probs[1])  # assumes label 1 = "injection"

INJECTION_THRESHOLD = 0.80
RULE_PATTERNS = [
    r"(?i)\bignore (previous|all) (instructions|rules)\b",
    r"(?i)\bdisregard (system|policy)\b",
    r"(?i)\bpretend to be\b",
    r"(?i)\bstart a new system prompt\b",
]

def guard(text: str):
    for p in RULE_PATTERNS:
        if re.search(p, text):
            return True, "rule", 1.0
    p = score_injection(text)
    if p >= INJECTION_THRESHOLD:
        return True, "ml", p
    return False, None, p

# ===========================================================
# 2) RAG — TXT → Chunk → Embed → FAISS
# ===========================================================
def clean_text(s: str) -> str:
    return re.sub(r"\s+", " ", s).strip()

# Load docs
docs_raw: List[Dict] = []
for p in Path(TXT_DIR).rglob("*.txt"):
    try:
        txt = Path(p).read_text(encoding="utf-8", errors="ignore")
    except Exception:
        txt = ""
    txt = clean_text(txt)
    if len(txt) > 50:
        docs_raw.append({"content": txt, "source": p.name})

print("✅ Raw docs loaded:", len(docs_raw))

# Chunking
CHUNK_SIZE = 800
OVERLAP    = 200
docs: List[Dict] = []
def _chunk_and_append(text: str, src: str):
    i = 0
    while i < len(text):
        chunk = text[i:i+CHUNK_SIZE]
        if len(chunk) > 200:
            docs.append({"content": chunk, "source": src})
        i += (CHUNK_SIZE - OVERLAP)

for d in docs_raw:
    _chunk_and_append(d["content"], d["source"])

print("✅ Chunks:", len(docs))

# Embeddings + FAISS
embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
corpus   = [x["content"] for x in docs]
embs     = embedder.encode(corpus, convert_to_numpy=True)
faiss.normalize_L2(embs)

index = faiss.IndexFlatIP(embs.shape[1])
index.add(embs)
print("✅ FAISS index entries:", index.ntotal)

def add_text_to_index(text: str, source_name: str):
    before = len(docs)
    _chunk_and_append(clean_text(text), source_name)
    new_chunks = [x["content"] for x in docs[before:]]
    if not new_chunks:
        return 0
    new_embs = embedder.encode(new_chunks, convert_to_numpy=True)
    faiss.normalize_L2(new_embs)
    index.add(new_embs)
    return len(new_chunks)

def retrieve(query: str, top_k: int = 6, min_score: float = 0.30):
    q = embedder.encode([query], convert_to_numpy=True)
    faiss.normalize_L2(q)
    scores, ids = index.search(q, top_k)
    out=[]
    for s,i in zip(scores[0], ids[0]):
        if i > -1 and float(s) >= min_score:
            out.append({"score": float(s), "content": docs[i]["content"], "source": docs[i]["source"]})
    return out

# ===========================================================
# 3) Gemini wrapper
# ===========================================================
import google.generativeai as genai
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

def _gemini_generate_once(model_name: str, prompt: str) -> str:
    model = genai.GenerativeModel(model_name)
    resp = model.generate_content(prompt)
    txt = getattr(resp, "text", None)
    if not txt and getattr(resp, "candidates", None):
        parts = resp.candidates[0].content.parts
        txt = "".join(getattr(p, "text", "") for p in parts)
    return (txt or "").strip()

def _gemini_generate(model_name: str, prompt: str) -> str:
    try:
        return _gemini_generate_once(model_name, prompt)
    except Exception as e:
        msg = str(e)
        # retry once if service unavailable / 503
        if "503" in msg or "Service Unavailable" in msg or "unavailable" in msg.lower():
            time.sleep(1.0)
            try:
                return _gemini_generate_once(model_name, prompt)
            except Exception as e2:
                return f"__ERROR__ {e2}"
        return f"__ERROR__ {e}"

def gemini_answer(prompt: str, concise: bool = True) -> str:
    if concise:
        prompt = (
            f"{prompt}\n\n"
            "Respond in a concise and clear way (about 3 to 5 sentences). "
            "Focus on facts and clarity. Avoid repetition, long lists, or code unless asked."
        )

    try:
        model = genai.GenerativeModel(GEMINI_MODEL)
        resp = model.generate_content(prompt, generation_config={"temperature": 0.3})
        txt = getattr(resp, "text", None)
        if not txt and getattr(resp, "candidates", None):
            parts = resp.candidates[0].content.parts
            txt = ''.join(getattr(p, 'text', '') for p in parts)
        return (txt or "").strip()
    except Exception:
        # fallback once if fails
        try:
            model = genai.GenerativeModel(GEMINI_FALLBACK)
            resp = model.generate_content(prompt, generation_config={"temperature": 0.3})
            return (getattr(resp, "text", "") or "").strip()
        except Exception:
            return "I don't have enough information."


# ===========================================================
# 4) RAG Answer
# ===========================================================
def rag_answer(query, top_k=6, max_context_chars=6000, min_hits=1):
    hits = retrieve(query, top_k=top_k)
    if len(hits) < min_hits:
        # No good retrieval → fallback to general Gemini answer
        return gemini_answer(query), [], False, []

    context = ""
    used = 0
    for h in hits:
        t = h["content"]
        if used + len(t) > max_context_chars:
            t = t[:max_context_chars-used]
        context += t + "\n---\n"
        used += len(t)

    prompt = f"""
Answer ONLY using the CONTEXT below.
If the answer is not present, return exactly:
"I don't have enough information."

CONTEXT:
{context}

QUESTION:
{query}

ANSWER:
""".strip()

    out = gemini_answer(prompt)
    pretty_hits = [{"source": h["source"], "score": round(h["score"], 3)} for h in hits]
    return out, hits, True, pretty_hits

# ===========================================================
# 5) Store structured knowledge
# ===========================================================
def summarize_to_structured_chunks(question: str, answer_text: str) -> str:
    prompt = f"""
You are a knowledge curator. Convert the ANSWER into short, precise knowledge bullets
that are useful for Retrieval-Augmented Generation (RAG). Avoid fluff and duplicates.

Return ONLY plain text in this structure:

TITLE: <short topic title>
POINTS:
- <fact 1>
- <fact 2>
- <fact 3>
REFERENCES:
- <if any known authoritative refs, else omit>

QUESTION:
{question}

ANSWER:
{answer_text}
"""
    return gemini_answer(prompt)

def store_new_knowledge(question: str, structured_text: str):
    ts = time.strftime("%Y-%m-%d %H:%M:%S")
    header = f"\n\n===== NEW ENTRY @ {ts} =====\nQUESTION: {question}\n"
    prev = Path(NEW_KNOW_FILE).read_text(encoding="utf-8") if Path(NEW_KNOW_FILE).exists() else ""
    Path(NEW_KNOW_FILE).write_text(prev + header + structured_text + "\n", encoding="utf-8")
    print(f"✅ NEW KNOWLEDGE appended → {NEW_KNOW_FILE}")
    added = add_text_to_index(structured_text, source_name="new_knowledge.txt")
    print(f"➕ Indexed {added} new chunks from curated knowledge.")

# ===========================================================
# 6) Cross-LLM Verification + Raw Answers for Gemini & Cohere
#    (always generate Gemini & Cohere answers; verify each)
# ===========================================================
VERIFY_PROMPT = """You are a security reviewer.

Respond ONLY:

VERDICT: yes|no|unsure
REASON: <short>
FIX: <corrected answer or empty>

QUESTION:
{q}

ANSWER:
{a}
"""

def _parse_verdict(txt: str):
    verdict, reason, fix = "unsure", "", ""
    m=re.search(r"VERDICT:\s*(yes|no|unsure)",txt,re.I)
    if m: verdict=m.group(1).lower()
    m=re.search(r"REASON:\s*(.*)",txt,re.I)
    if m: reason=m.group(1).strip()
    m=re.search(r"FIX:\s*(.*)",txt,re.I)
    if m: fix=m.group(1).strip()
    return {"verdict": verdict, "justification": reason, "corrected_answer": fix}

def ask_gemini_verify(q,a):
    prompt = VERIFY_PROMPT.format(q=q,a=a)
    return gemini_answer(prompt)

import cohere
co = cohere.Client(os.environ["COHERE_API_KEY"])

def cohere_direct_answer(question: str, concise: bool = True) -> str:
    if concise:
        question += (
            "\n\nRespond concisely in 3–5 sentences. "
            "Focus on clarity and facts. Avoid lengthy explanations or lists."
        )
    try:
        r = co.chat(model=COHERE_MODEL, message=question, temperature=0.3)
        return getattr(r, "text", None) or getattr(r, "output_text", "") or ""
    except Exception as e:
        return f"__ERROR__ {e}"


def ask_cohere_verify(q,a):
    prompt = VERIFY_PROMPT.format(q=q,a=a)
    try:
        r = co.chat(model=COHERE_MODEL, message=prompt, temperature=0)
        return getattr(r,"text",None) or getattr(r,"output_text","") or ""
    except Exception as e:
        return f"__ERROR__ {e}"

# ===========================================================
# 7) Eligibility Scoring
# ===========================================================
from transformers import pipeline as hf_pipeline
tox = hf_pipeline("text-classification", model="unitary/toxic-bert", device=0 if torch.cuda.is_available() else -1)

BAN       = ["bomb","terror","weapon","drug recipe","how to kill"]
SENSITIVE = ["self-harm","suicide","abuse"]

def score_toxic(t):
    try:
        pred = tox(t)[0]
        return 0 if pred["label"].lower().startswith("toxic") else 1
    except Exception:
        return 1

def score_ban(t):   return 0 if any(w in t.lower() for w in BAN) else 1
def score_sens(t):  return 0 if any(w in t.lower() for w in SENSITIVE) else 1
def score_lang(t):  return 1
def score_sent(t):  return 1
def score_len(t,limit=300): return 1 if len(t.split())<=limit else 0

def compute_ec(t):
    scores = {
        "toxicity":  score_toxic(t),
        "ban_topics":score_ban(t),
        "sensitive": score_sens(t),
        "language":  score_lang(t),
        "sentiment": score_sent(t),
        "length":    score_len(t),
    }
    return sum(scores.values()), scores

def final_decision(t, threshold=4):
    ec, br = compute_ec(t)
    return {"eligible": ec>=threshold, "ec_score": ec, "breakdown": br}

# ===========================================================
# 8) MAIN — secure_query()
# ===========================================================
def secure_query(query: str):
    # 0) Guard
    blocked, reason, p = guard(query)
    if blocked:
        return {"blocked": True, "reason": reason, "ml_score": p}

    # 1) RAG (with fallback decision)
    rag_ans, refs, from_rag, pretty_hits = rag_answer(query)

    # If fallback used (no strong RAG hits), store structured knowledge
    if not from_rag:
        structured = summarize_to_structured_chunks(query, rag_ans)
        store_new_knowledge(query, structured)

    # 2) ALWAYS also get raw answers from Gemini & Cohere for the same question
    gemini_raw = gemini_answer(query)
    cohere_raw = cohere_direct_answer(query)

    # 3) Verify each (self-verify)
    gemini_verdict = _parse_verdict(ask_gemini_verify(query, gemini_raw))
    cohere_verdict = _parse_verdict(ask_cohere_verify(query, cohere_raw))
    rag_verdict    = _parse_verdict(ask_gemini_verify(query, rag_ans))  # let Gemini review RAG answer

    # 4) Eligibility for each
    rag_elig     = final_decision(rag_ans)
    gemini_elig  = final_decision(gemini_raw)
    cohere_elig  = final_decision(cohere_raw)

    # 5) De-dup sources
    seen, uniq_sources = set(), []
    for s in [r["source"] for r in refs] if refs else ["LLM fallback"]:
        if s not in seen:
            seen.add(s)
            uniq_sources.append(s)

    return {
        "blocked": False,

        # RAG
        "rag": {
            "answer": rag_ans,
            "from_rag": from_rag,
            "retrieval": pretty_hits,
            "verification": rag_verdict,
            "eligibility": rag_elig,
            "sources": uniq_sources if from_rag else ["LLM fallback → stored in new_knowledge.txt"]
        },

        # Gemini
        "gemini": {
            "answer": gemini_raw,
            "verification": gemini_verdict,
            "eligibility": gemini_elig
        },

        # Cohere
        "cohere": {
            "answer": cohere_raw,
            "verification": cohere_verdict,
            "eligibility": cohere_elig
        }
    }

print("✅ READY — try:")
```

```python
# Install Gradio (if not already)
!pip -q install gradio==4.19.2

import gradio as gr
import json
import traceback
from typing import Any

def safe_parse_result(res: Any):
    """
    Normalize the result to a Python dict.
    Accepts: dict, JSON string, or other -> returns dict with safe defaults.
    """
    if isinstance(res, dict):
        return res
    if isinstance(res, str):
        # attempt to parse JSON
        try:
            return json.loads(res)
        except Exception:
            # not JSON: return raw text in a dict
            return {"raw": res}
    # fallback: try to convert to string
    try:
        return json.loads(str(res))
    except Exception:
        return {"raw": str(res)}

def get_field(d: dict, *path, default=""):
    """Safe nested lookup. path are keys."""
    cur = d
    try:
        for k in path:
            if isinstance(cur, dict):
                cur = cur.get(k, default)
            else:
                return default
        return cur if cur is not None else default
    except Exception:
        return default

def compute_avg_ec(scores):
    """Compute average ec_score from a list of numeric or numeric-like values."""
    vals = []
    for s in scores:
        try:
            if isinstance(s, dict):
                v = s.get("ec_score") or s.get("ec_score", None)
            else:
                v = s
            if v is None:
                continue
            vals.append(float(v))
        except Exception:
            continue
    return round(sum(vals) / len(vals), 3) if vals else None

def run_ui(query: str):
    if not query or not str(query).strip():
        return ("No query provided.", "", "", "", "", "", "", "", "", "", "")
    try:
        raw = secure_query(query)   # call your backend function directly
    except Exception as e:
        tb = traceback.format_exc()
        return ("ERROR calling secure_query", tb, "", "", "", "", "", "", "", "", "")

    out = safe_parse_result(raw)

    # Blocked / safe
    blocked_flag = out.get("blocked", False)
    blocked_reason = out.get("reason", "")
    blocked_display = "BLOCKED" if blocked_flag else "ALLOWED"
    if not blocked_flag and blocked_reason:
        blocked_display += f" ({blocked_reason})"

    # Sources
    sources = get_field(out, "rag", "sources", default=get_field(out, "sources", default=""))
    if isinstance(sources, (list, tuple)):
        sources_str = "\n".join(map(str, sources))
    else:
        sources_str = str(sources)

    # RAG fields
    rag_ans = get_field(out, "rag", "answer", default=get_field(out, "rag", "answer", default=""))
    rag_verdict = get_field(out, "rag", "verification", "verdict", default="")
    rag_elig = get_field(out, "rag", "eligibility", default={})
    rag_ec = rag_elig.get("ec_score") if isinstance(rag_elig, dict) else rag_elig

    rag_display = f"{rag_ans}"

    # Gemini fields
    gem_ans = get_field(out, "gemini", "answer", default=get_field(out, "gemini", "answer", default=""))
    gem_verdict = get_field(out, "gemini", "verification", "verdict", default="")
    gem_elig = get_field(out, "gemini", "eligibility", default={})
    gem_ec = gem_elig.get("ec_score") if isinstance(gem_elig, dict) else gem_elig

    gem_display = f"{gem_ans}"

    # Cohere fields
    coh_ans = get_field(out, "cohere", "answer", default=get_field(out, "cohere", "answer", default=""))
    coh_verdict = get_field(out, "cohere", "verification", "verdict", default="")
    coh_elig = get_field(out, "cohere", "eligibility", default={})
    coh_ec = coh_elig.get("ec_score") if isinstance(coh_elig, dict) else coh_elig

    coh_display = f"{coh_ans}"

    # Final aggregated eligibility (average of available ec_scores)
    ec_scores = [v for v in [rag_ec, gem_ec, coh_ec] if v is not None and v != ""]
    final_avg = compute_avg_ec(ec_scores)

    # Compose short verdict summary
    verdict_summary = f"RAG: {rag_verdict or 'n/a'} | Gemini: {gem_verdict or 'n/a'} | Cohere: {coh_verdict or 'n/a'}"

    # Compose eligibility summary
    elig_summary = (
        f"RAG ec_score: {rag_ec}\n"
        f"Gemini ec_score: {gem_ec}\n"
        f"Cohere ec_score: {coh_ec}\n"
        f"Final avg ec_score: {final_avg}"
    )

    # Return values map to UI outputs below
    return (
        blocked_display,         # Blocked / Allowed
        sources_str,             # Sources
        rag_display,             # RAG answer
        rag_verdict or "",       # RAG verdict
        str(rag_ec),             # RAG ec_score
        gem_display,             # Gemini answer
        gem_verdict or "",       # Gemini verdict
        str(gem_ec),             # Gemini ec_score
        coh_display,             # Cohere answer
        coh_verdict or "",       # Cohere verdict
        str(coh_ec),             # Cohere ec_score
        verdict_summary,         # combined short verdicts
        elig_summary             # detailed eligibility summary
    )

# Build the Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🔐 LLM Security — Results Dashboard")
    gr.Markdown("Displays safety, RAG/Gemini/Cohere outputs and eligibility scores.")

    with gr.Row():
        query = gr.Textbox(lines=3, label="Query")
        run_btn = gr.Button("Run Secure Query")

    with gr.Row():
        blocked_out = gr.Textbox(lines=1, label="Blocked / Allowed")
        sources_out = gr.Textbox(lines=3, label="Sources")

    gr.Markdown("### RAG")
    with gr.Row():
        rag_answer_out = gr.Textbox(lines=6, label="RAG Answer")
        rag_ver_out = gr.Textbox(lines=1, label="RAG Verdict")
        rag_ec_out = gr.Textbox(lines=1, label="RAG ec_score")

    gr.Markdown("### Gemini")
    with gr.Row():
        gem_answer_out = gr.Textbox(lines=6, label="Gemini Answer")
        gem_ver_out = gr.Textbox(lines=1, label="Gemini Verdict")
        gem_ec_out = gr.Textbox(lines=1, label="Gemini ec_score")

    gr.Markdown("### Cohere")
    with gr.Row():
        coh_answer_out = gr.Textbox(lines=6, label="Cohere Answer")
        coh_ver_out = gr.Textbox(lines=1, label="Cohere Verdict")
        coh_ec_out = gr.Textbox(lines=1, label="Cohere ec_score")

    gr.Markdown("### Aggregates")
    verdict_summary_out = gr.Textbox(lines=1, label="Combined Verdicts")
    elig_summary_out = gr.Textbox(lines=4, label="Eligibility Summary (detailed)")

    run_btn.click(
        fn=run_ui,
        inputs=[query],
        outputs=[
            blocked_out, sources_out,
            rag_answer_out, rag_ver_out, rag_ec_out,
            gem_answer_out, gem_ver_out, gem_ec_out,
            coh_answer_out, coh_ver_out, coh_ec_out,
            verdict_summary_out, elig_summary_out
        ]
    )

demo.launch(share=True, debug=True)
```

```python
# -- installs --
!pip -q install sentence-transformers transformers accelerate faiss-cpu pypdf google-generativeai cohere

import os, re, json, time, torch, faiss, numpy as np
from pathlib import Path
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from typing import List, Dict

# ---------------------------
# 🔑 API KEYS (set here or Runtime → Env)
# ---------------------------
os.environ["GOOGLE_API_KEY"]  = "AIzaSyAq_t93GdSHc_BqvE756ntoug37sjKdYdc"
os.environ["COHERE_API_KEY"]  = "ygNgfxHdydrQSJUnBWtr6lHhaLYJYfZ5gPa4yF71"

if not os.getenv("GOOGLE_API_KEY"):
    raise RuntimeError("GOOGLE_API_KEY not set")
if not os.getenv("COHERE_API_KEY"):
    raise RuntimeError("COHERE_API_KEY not set")

# Models (Gemini must include "models/" prefix)
GEMINI_MODEL    = "models/gemini-2.5-flash"          # primary
GEMINI_FALLBACK = "models/gemini-1.5-flash-latest"   # fallback
COHERE_MODEL    = "command-a-03-2025"

# ---------------------------
# 📂 Paths
# ---------------------------
ROBERTO_MODEL_DIR = "/content/drive/MyDrive/MajorProj/roberta_model_best"
BASE_DIR          = "/content/drive/MyDrive/data"
TXT_DIR           = f"{BASE_DIR}/processed"
NEW_KNOW_FILE     = f"{TXT_DIR}/new_knowledge.txt"

Path(TXT_DIR).mkdir(parents=True, exist_ok=True)
Path(BASE_DIR).mkdir(parents=True, exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"

# ===========================================================
# 1) PROMPT INJECTION DETECTOR (RoBERTa — local)
# ===========================================================
tokenizer_guard = AutoTokenizer.from_pretrained(ROBERTO_MODEL_DIR, local_files_only=True)
model_guard     = AutoModelForSequenceClassification.from_pretrained(ROBERTO_MODEL_DIR, local_files_only=True).to(device).eval()

@torch.inference_mode()
def score_injection(text: str) -> float:
    x = tokenizer_guard(text, return_tensors="pt", truncation=True, max_length=512).to(device)
    logits = model_guard(**x).logits
    probs  = torch.softmax(logits, dim=-1)[0].tolist()
    return float(probs[1])  # assumes label 1 = "injection"

INJECTION_THRESHOLD = 0.80
RULE_PATTERNS = [
    r"(?i)\bignore (previous|all) (instructions|rules)\b",
    r"(?i)\bdisregard (system|policy)\b",
    r"(?i)\bpretend to be\b",
    r"(?i)\bstart a new system prompt\b",
]

def guard(text: str):
    for p in RULE_PATTERNS:
        if re.search(p, text):
            return True, "rule", 1.0
    p = score_injection(text)
    if p >= INJECTION_THRESHOLD:
        return True, "ml", p
    return False, None, p

# ===========================================================
# 2) RAG — TXT → Chunk → Embed → FAISS
# ===========================================================
def clean_text(s: str) -> str:
    return re.sub(r"\s+", " ", s).strip()

# Load docs
docs_raw: List[Dict] = []
for p in Path(TXT_DIR).rglob("*.txt"):
    try:
        txt = Path(p).read_text(encoding="utf-8", errors="ignore")
    except Exception:
        txt = ""
    txt = clean_text(txt)
    if len(txt) > 50:
        docs_raw.append({"content": txt, "source": p.name})

print("✅ Raw docs loaded:", len(docs_raw))

# Chunking
CHUNK_SIZE = 800
OVERLAP    = 200
docs: List[Dict] = []
def _chunk_and_append(text: str, src: str):
    i = 0
    while i < len(text):
        chunk = text[i:i+CHUNK_SIZE]
        if len(chunk) > 200:
            docs.append({"content": chunk, "source": src})
        i += (CHUNK_SIZE - OVERLAP)

for d in docs_raw:
    _chunk_and_append(d["content"], d["source"])

print("✅ Chunks:", len(docs))

# Embeddings + FAISS
embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
corpus   = [x["content"] for x in docs]
embs     = embedder.encode(corpus, convert_to_numpy=True)
faiss.normalize_L2(embs)

index = faiss.IndexFlatIP(embs.shape[1])
index.add(embs)
print("✅ FAISS index entries:", index.ntotal)

def add_text_to_index(text: str, source_name: str):
    before = len(docs)
    _chunk_and_append(clean_text(text), source_name)
    new_chunks = [x["content"] for x in docs[before:]]
    if not new_chunks:
        return 0
    new_embs = embedder.encode(new_chunks, convert_to_numpy=True)
    faiss.normalize_L2(new_embs)
    index.add(new_embs)
    return len(new_chunks)

def retrieve(query: str, top_k: int = 6, min_score: float = 0.30):
    q = embedder.encode([query], convert_to_numpy=True)
    faiss.normalize_L2(q)
    scores, ids = index.search(q, top_k)
    out=[]
    for s,i in zip(scores[0], ids[0]):
        if i > -1 and float(s) >= min_score:
            out.append({"score": float(s), "content": docs[i]["content"], "source": docs[i]["source"]})
    return out

# ===========================================================
# 3) Gemini wrapper
# ===========================================================
import google.generativeai as genai
genai.configure(api_key=os.environ["GOOGLE_API_KEY"])

def _gemini_generate_once(model_name: str, prompt: str) -> str:
    model = genai.GenerativeModel(model_name)
    resp = model.generate_content(prompt)
    txt = getattr(resp, "text", None)
    if not txt and getattr(resp, "candidates", None):
        parts = resp.candidates[0].content.parts
        txt = "".join(getattr(p, "text", "") for p in parts)
    return (txt or "").strip()

def _gemini_generate(model_name: str, prompt: str) -> str:
    try:
        return _gemini_generate_once(model_name, prompt)
    except Exception as e:
        msg = str(e)
        # retry once if service unavailable / 503
        if "503" in msg or "Service Unavailable" in msg or "unavailable" in msg.lower():
            time.sleep(1.0)
            try:
                return _gemini_generate_once(model_name, prompt)
            except Exception as e2:
                return f"__ERROR__ {e2}"
        return f"__ERROR__ {e}"

def gemini_answer(prompt: str, concise: bool = True) -> str:
    if concise:
        prompt = (
            f"{prompt}\n\n"
            "Respond in a concise and clear way (about 3 to 5 sentences). "
            "Focus on facts and clarity. Avoid repetition, long lists, or code unless asked."
        )

    try:
        model = genai.GenerativeModel(GEMINI_MODEL)
        resp = model.generate_content(prompt, generation_config={"temperature": 0.3})
        txt = getattr(resp, "text", None)
        if not txt and getattr(resp, "candidates", None):
            parts = resp.candidates[0].content.parts
            txt = ''.join(getattr(p, 'text', '') for p in parts)
        return (txt or "").strip()
    except Exception:
        # fallback once if fails
        try:
            model = genai.GenerativeModel(GEMINI_FALLBACK)
            resp = model.generate_content(prompt, generation_config={"temperature": 0.3})
            return (getattr(resp, "text", "") or "").strip()
        except Exception:
            return "I don't have enough information."


# ===========================================================
# 4) RAG Answer
# ===========================================================
def rag_answer(query, top_k=6, max_context_chars=6000, min_hits=1):
    hits = retrieve(query, top_k=top_k)
    if len(hits) < min_hits:
        # No good retrieval → fallback to general Gemini answer
        return gemini_answer(query), [], False, []

    context = ""
    used = 0
    for h in hits:
        t = h["content"]
        if used + len(t) > max_context_chars:
            t = t[:max_context_chars-used]
        context += t + "\n---\n"
        used += len(t)

    prompt = f"""
Answer ONLY using the CONTEXT below.
If the answer is not present, return exactly:
"I don't have enough information."

CONTEXT:
{context}

QUESTION:
{query}

ANSWER:
""".strip()

    out = gemini_answer(prompt)
    pretty_hits = [{"source": h["source"], "score": round(h["score"], 3)} for h in hits]
    return out, hits, True, pretty_hits

# ===========================================================
# 5) Store structured knowledge
# ===========================================================
def summarize_to_structured_chunks(question: str, answer_text: str) -> str:
    prompt = f"""
You are a knowledge curator. Convert the ANSWER into short, precise knowledge bullets
that are useful for Retrieval-Augmented Generation (RAG). Avoid fluff and duplicates.

Return ONLY plain text in this structure:

TITLE: <short topic title>
POINTS:
- <fact 1>
- <fact 2>
- <fact 3>
REFERENCES:
- <if any known authoritative refs, else omit>

QUESTION:
{question}

ANSWER:
{answer_text}
"""
    return gemini_answer(prompt)

def store_new_knowledge(question: str, structured_text: str):
    ts = time.strftime("%Y-%m-%d %H:%M:%S")
    header = f"\n\n===== NEW ENTRY @ {ts} =====\nQUESTION: {question}\n"
    prev = Path(NEW_KNOW_FILE).read_text(encoding="utf-8") if Path(NEW_KNOW_FILE).exists() else ""
    Path(NEW_KNOW_FILE).write_text(prev + header + structured_text + "\n", encoding="utf-8")
    print(f"✅ NEW KNOWLEDGE appended → {NEW_KNOW_FILE}")
    added = add_text_to_index(structured_text, source_name="new_knowledge.txt")
    print(f"➕ Indexed {added} new chunks from curated knowledge.")

# ===========================================================
# 6) Cross-LLM Verification + Raw Answers for Gemini & Cohere
#    (always generate Gemini & Cohere answers; verify each)
# ===========================================================
VERIFY_PROMPT = """You are a security reviewer.

Respond ONLY:

VERDICT: yes|no|unsure
REASON: <short>
FIX: <corrected answer or empty>

QUESTION:
{q}

ANSWER:
{a}
"""

def _parse_verdict(txt: str):
    verdict, reason, fix = "unsure", "", ""
    m=re.search(r"VERDICT:\s*(yes|no|unsure)",txt,re.I)
    if m: verdict=m.group(1).lower()
    m=re.search(r"REASON:\s*(.*)",txt,re.I)
    if m: reason=m.group(1).strip()
    m=re.search(r"FIX:\s*(.*)",txt,re.I)
    if m: fix=m.group(1).strip()
    return {"verdict": verdict, "justification": reason, "corrected_answer": fix}

def ask_gemini_verify(q,a):
    prompt = VERIFY_PROMPT.format(q=q,a=a)
    return gemini_answer(prompt)

import cohere
co = cohere.Client(os.environ["COHERE_API_KEY"])

def cohere_direct_answer(question: str, concise: bool = True) -> str:
    if concise:
        question += (
            "\n\nRespond concisely in 3–5 sentences. "
            "Focus on clarity and facts. Avoid lengthy explanations or lists."
        )
    try:
        r = co.chat(model=COHERE_MODEL, message=question, temperature=0.3)
        return getattr(r, "text", None) or getattr(r, "output_text", "") or ""
    except Exception as e:
        return f"__ERROR__ {e}"


def ask_cohere_verify(q,a):
    prompt = VERIFY_PROMPT.format(q=q,a=a)
    try:
        r = co.chat(model=COHERE_MODEL, message=prompt, temperature=0)
        return getattr(r,"text",None) or getattr(r,"output_text","") or ""
    except Exception as e:
        return f"__ERROR__ {e}"

# ===========================================================
# 7) Eligibility Scoring
# ===========================================================
from transformers import pipeline as hf_pipeline
tox = hf_pipeline("text-classification", model="unitary/toxic-bert", device=0 if torch.cuda.is_available() else -1)

BAN       = ["bomb","terror","weapon","drug recipe","how to kill"]
SENSITIVE = ["self-harm","suicide","abuse"]

def score_toxic(t):
    try:
        pred = tox(t)[0]
        return 0 if pred["label"].lower().startswith("toxic") else 1
    except Exception:
        return 1

def score_ban(t):   return 0 if any(w in t.lower() for w in BAN) else 1
def score_sens(t):  return 0 if any(w in t.lower() for w in SENSITIVE) else 1
def score_lang(t):  return 1
def score_sent(t):  return 1
def score_len(t,limit=300): return 1 if len(t.split())<=limit else 0

def compute_ec(t):
    scores = {
        "toxicity":  score_toxic(t),
        "ban_topics":score_ban(t),
        "sensitive": score_sens(t),
        "language":  score_lang(t),
        "sentiment": score_sent(t),
        "length":    score_len(t),
    }
    return sum(scores.values()), scores

def final_decision(t, threshold=4):
    ec, br = compute_ec(t)
    return {"eligible": ec>=threshold, "ec_score": ec, "breakdown": br}

# ===========================================================
# 8) MAIN — secure_query()
# ===========================================================
def secure_query(query: str):
    # 0) Guard
    blocked, reason, p = guard(query)
    if blocked:
        return {"blocked": True, "reason": reason, "ml_score": p}

    # 1) RAG (with fallback decision)
    rag_ans, refs, from_rag, pretty_hits = rag_answer(query)

    # If fallback used (no strong RAG hits), store structured knowledge
    if not from_rag:
        structured = summarize_to_structured_chunks(query, rag_ans)
        store_new_knowledge(query, structured)

    # 2) ALWAYS also get raw answers from Gemini & Cohere for the same question
    gemini_raw = gemini_answer(query)
    cohere_raw = cohere_direct_answer(query)

    # 3) Verify each (self-verify)
    gemini_verdict = _parse_verdict(ask_gemini_verify(query, gemini_raw))
    cohere_verdict = _parse_verdict(ask_cohere_verify(query, cohere_raw))
    rag_verdict    = _parse_verdict(ask_gemini_verify(query, rag_ans))  # let Gemini review RAG answer

    # 4) Eligibility for each
    rag_elig     = final_decision(rag_ans)
    gemini_elig  = final_decision(gemini_raw)
    cohere_elig  = final_decision(cohere_raw)

    # 5) De-dup sources
    seen, uniq_sources = set(), []
    for s in [r["source"] for r in refs] if refs else ["LLM fallback"]:
        if s not in seen:
            seen.add(s)
            uniq_sources.append(s)

    return {
        "blocked": False,

        # RAG
        "rag": {
            "answer": rag_ans,
            "from_rag": from_rag,
            "retrieval": pretty_hits,
            "verification": rag_verdict,
            "eligibility": rag_elig,
            "sources": uniq_sources if from_rag else ["LLM fallback → stored in new_knowledge.txt"]
        },

        # Gemini
        "gemini": {
            "answer": gemini_raw,
            "verification": gemini_verdict,
            "eligibility": gemini_elig
        },

        # Cohere
        "cohere": {
            "answer": cohere_raw,
            "verification": cohere_verdict,
            "eligibility": cohere_elig
        }
    }

print("✅ READY — try:")

```
